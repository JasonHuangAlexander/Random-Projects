```python
import csv
import numpy as np
x = [] #features
y = [] #target values
file = open("avocado.csv","r")
reader = csv.reader(file)
for row in reader:
    if not (row[1]=="Date"):
        x.append(float(str(row[1])[5:7]))
    if not (row[2]=="AveragePrice"):
        y.append(float(row[2]))
x = np.array(x)
y = np.array(y)
w = 0
b = 0
def gradients(x, y, w, b):
    d_dw = 0
    d_db = 0
    for i in range(x.shape[0]):
        d_dw_i = (w*x[i]+b-y[i])*x[i]
        d_db_i = (w*x[i]+b-y[i])
        d_db = d_db + d_db_i
        d_dw = d_dw + d_dw_i
    d_dw = d_dw/x.shape[0]
    d_db = d_db/x.shape[0]
    return d_dw, d_db

def gradientDescent(x, y, w, b, alpha):
    iterations = 1000
    for i in range(iterations):
        d_dw, d_db = gradients(x, y, w ,b)
        w = float(w - alpha * d_dw)
        b = float(b - alpha * d_db)
        print("iteration " + str(i) + ": w = " + str(w) + "  b = " + str(b))
    return w, b

inp = input("Enter a month(1-12) to predict avocado price: ")
w, b = gradientDescent(x, y, w, b, 0.01)
prediction = w*float(inp) + b
print("Predicted price: ")
print(float(prediction))
```
